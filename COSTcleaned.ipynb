```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "Retail_Dataset(Sheet1).csv"
df = pd.read_csv("Retail_Dataset(Sheet1).csv", encoding="latin1")
df.head()
```

```python
cols = ["rating_before", "rating_after", "price_target_before", "price_target_after"]
df = df[df["ticker"] == "COST"].copy()
df[cols] = df[cols].replace(["NOT FOUND", "", " "], np.nan)
df[cols] = df[cols].apply(lambda col: col.map(lambda x: np.nan if isinstance(x, str) and x.strip() == "" else x))
cost_clean = df.dropna(subset=cols)
cost_clean.head(20)
```

```python
cost_clean = cost_clean.rename(columns={"date" : "Date", "broker" : "Broker", "company_Name" : "Company Name", "ticker" : "Ticker", "analytst": "Analyst", 
                                    "rating_before" : "Rating Before", "rating_after" : "Rating After", "price_target_before" : "Price Target Before", 
                                    "price_target_after" : "Price Target After"})
cols_to_title = ["Broker", "Analyst", "Rating Before", "Rating After"]
for col in cols_to_title:
    cost_clean[col] = cost_clean[col].str.title()
```

```python
cost_clean["Date"] = pd.to_datetime(cost_clean["Date"], errors="coerce")
cost_clean = cost_clean.sort_values(by="Date")
cost_clean.head(30)
```
