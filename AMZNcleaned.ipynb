```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "Retail_Dataset(Sheet1).csv"
am_df = pd.read_csv("Retail_Dataset(Sheet1).csv", encoding = "latin1")
am_df.head()
```

```python
cols = ["rating_before", "rating_after", "price_target_before", "price_target_after"]
am_df = am_df[am_df["ticker"] == "AMZN"].copy()
am_df[cols] = am_df[cols].replace(["NOT FOUND", "", " "], np.nan)
am_df[cols] = am_df[cols].apply(lambda col: col.map(lambda x: np.nan if isinstance(x, str) and x.strip() == "" else x))
am_clean = am_df.dropna(subset=cols)
am_clean.head(20)
```

```python
am_clean = am_clean.rename(columns={"date" : "Date", "broker" : "Broker", "company_Name" : "Company Name", "ticker" : "Ticker", "analytst": "Analyst", 
                                    "rating_before" : "Rating Before", "rating_after" : "Rating After", "price_target_before" : "Price Target Before", 
                                    "price_target_after" : "Price Target After"})
cols_to_title = ["Broker", "Analyst", "Rating Before", "Rating After"]
for col in cols_to_title:
    am_clean[col] = am_clean[col].str.title()

am_clean["Date"] = pd.to_datetime(am_clean["Date"], errors="coerce")
am_clean = am_clean.sort_values(by="Date")
am_clean.head(30)
```

```python
am_clean.to_csv("AMZNcleaned.csv", index=False)
```
