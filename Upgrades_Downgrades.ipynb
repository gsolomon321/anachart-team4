```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "cleaned_dataset - Sheet1.csv"
clean = pd.read_csv("cleaned_dataset - Sheet1.csv", encoding = "latin1")
clean.head()
```

```python
rating_map = {"Buy": 1, "Outperform": 1, "Positive": 1, "Strong Buy": 1, "Hold": 0, "Neutral": 0, 
              "Market Perform": 0, "Sector Perform": 0, "Sell": -1, "Short": -1, "Underperform": -1}

clean['Rating Before Score'] = clean['Rating Before'].map(rating_map)
clean['Rating After Score'] = clean['Rating After'].map(rating_map)

clean.head()
```

```python
def classify_rating_change(row):
    if row['Rating After Score'] > row['Rating Before Score']:
        return 'Upgrade'
    elif row['Rating After Score'] < row['Rating Before Score']:
        return 'Downgrade'
    else:
        return 'No Change'
clean['Event Type'] = clean.apply(classify_rating_change, axis = 1)

clean = clean.drop(columns = ['Rating Before Score', 'Rating After Score', 'Upgrade/Downgrade/No Change'])
clean.head()
```

```python
clean['Event Type'].value_counts()
```

```python
upgrades = clean[clean['Event Type'] == 'Upgrade']
downgrades = clean[clean['Event Type'] == 'Downgrade']
no_change = clean[clean['Event Type'] == 'No Change']
```
