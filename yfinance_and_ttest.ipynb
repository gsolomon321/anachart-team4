# **Welch Two-Sample T-Test**

### **30-day returns**

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("t_test_30 - Sheet1 (1).csv")

df.head(40)
```

```python
upgrade_returns = df["Upgrade: 30-day returns"].dropna()
downgrade_returns = df["Downgrade: 30-day returns"].dropna()
```

```python
upgrade_returns.head(30)
```

```python
downgrade_returns.head(40)
```

```python
t_stat, p_value = stats.ttest_ind(upgrade_returns, downgrade_returns, equal_var=False)

print("Upgrade sample size:", len(upgrade_returns))
print("Downgrade sample size:", len(downgrade_returns))
print("Upgrade mean return:", upgrade_returns.mean())
print("Downgrade mean return:", downgrade_returns.mean())
print("Upgrade standard deviation:", upgrade_returns.std())
print("Downgrade standard deviation:", downgrade_returns.std())
print("Downgrade mean return:", downgrade_returns.mean())
print("Welch t-statistic:", t_stat)
print("p-value:", p_value)
```
