```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "cleaned_dataset - Sheet1.csv"
clean_df = pd.read_csv("cleaned_dataset - Sheet1.csv", encoding = "latin1")
clean_df.head()
```

```python
cols = ["rating_before", "rating_after", "price_target_before", "price_target_after"]
ro_df = ro_df[ro_df["ticker"] == "ROST"].copy()
ro_df[cols] = ro_df[cols].replace(["NOT FOUND", "", " "], np.nan)
ro_df[cols] = ro_df[cols].apply(lambda col: col.map(lambda x: np.nan if isinstance(x, str) and x.strip() == "" else x))
ro_clean = ro_df.dropna(subset=cols)
ro_clean.head(20)
```

```python
ro_clean = ro_clean.rename(columns={"date" : "Date", "broker" : "Broker", "company_Name" : "Company Name", "ticker" : "Ticker", "analytst": "Analyst", 
                                    "rating_before" : "Rating Before", "rating_after" : "Rating After", "price_target_before" : "Price Target Before", 
                                    "price_target_after" : "Price Target After"})
cols_to_title = ["Broker", "Analyst", "Rating Before", "Rating After"]
for col in cols_to_title:
    ro_clean[col] = ro_clean[col].str.title()

ro_clean["Date"] = pd.to_datetime(ro_clean["Date"], errors="coerce")
ro_clean = ro_clean.sort_values(by="Date")
ro_clean.head(30)
```

```python
ro_clean.to_csv("ROSTcleaned.csv", index=False)
```
