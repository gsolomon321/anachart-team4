```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "Retail_Dataset(Sheet1).csv"
lu_df = pd.read_csv("Retail_Dataset(Sheet1).csv", encoding = "latin1")
lu_df.head()
```

```python
cols = ["rating_before", "rating_after", "price_target_before", "price_target_after"]
lu_df = lu_df[lu_df["ticker"] == "LULU"].copy()
lu_df[cols] = lu_df[cols].replace(["NOT FOUND", "", " "], np.nan)
lu_df[cols] = lu_df[cols].apply(lambda col: col.map(lambda x: np.nan if isinstance(x, str) and x.strip() == "" else x))
lu_clean = lu_df.dropna(subset=cols)
lu_clean.head(20)
```

```python
lu_clean = lu_clean.rename(columns={"date" : "Date", "broker" : "Broker", "company_Name" : "Company Name", "ticker" : "Ticker", "analytst": "Analyst", 
                                    "rating_before" : "Rating Before", "rating_after" : "Rating After", "price_target_before" : "Price Target Before", 
                                    "price_target_after" : "Price Target After"})
cols_to_title = ["Broker", "Analyst", "Rating Before", "Rating After"]
for col in cols_to_title:
    lu_clean[col] = lu_clean[col].str.title()

lu_clean["Date"] = pd.to_datetime(lu_clean["Date"], errors="coerce")
lu_clean = lu_clean.sort_values(by="Date")
lu_clean.head(30)
```

```python
lu_clean.to_csv("LULUcleaned.csv", index=False)
```
